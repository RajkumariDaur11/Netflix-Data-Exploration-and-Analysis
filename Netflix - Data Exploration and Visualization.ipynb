# Duration Insights

```python
df['duration'].unique()
```

```python
df['duration'].value_counts().head(60)
```

```python
Duration_Refined = df["duration"].value_counts()
Duration_Refined = Duration_Refined[Duration_Refined > 20].reset_index()
Duration_Refined.rename(columns={'index':'Duration','duration':'Counts'},inplace = True)
```

```python
sns.scatterplot(y = "Counts", x="Duration", data = Duration_Refined)
```

### Observations
The maximum duartion time lies in the range of 80-120 minutes

# Modification of Data

```python
Movies_genre = df.explode('genre')
Movies_stars = df.explode('stars')
Movies_votes = df.explode('votes')
```

```python
Movies_genre_refined = Movies_genre['genre'].value_counts()
Movies_genre_refined = Movies_genre_refined[Movies_genre_refined > 100].reset_index()
Movies_genre_refined.rename(columns={'index':'genre','genre':'Counts'}, inplace = True) 
sns.barplot(x="Counts", y ="genre" , data = Movies_genre_refined)
plt.show()
Movies_genre['genre'].value_counts()
```

#### Observation
Animation, action and adventure are most watched genre
